**Module 5 Assignment: Using Pandas to Analyze Diamonds Data**

(3 Points) What proportion of diamonds are between .30 and 1.08 carats?

(3 Points) How many of the diamonds have equal x and y dimensions?

(3 Points) How many of the diamonds have a carat less than the mean carat value?

(3 Points) How many diamonds have a Premium cut or better? Note that the possible values for the quality of a cut are ranked in ascending order as follows: Fair / Good / Very Good / Premium / Ideal

(3 Points) Which diamond has the highest price per carat? What is its value?

(3 Points) Make a bar plot of the ‘cut’ data. What does it tell us about the frequencies of the individual cut values within the data set?

(4 Points) Make boxplots of the diamond price for each cut and discuss any conclusions you can draw from the appearance of the boxplots.

(4 Points) Make histograms of the length, width, depth, and total depth attributes and discuss any conclusions you can draw from the shape of the resulting distributions.

(4 Points) Make a scatter plot of price vs. carat. What can we say about the relationship between those two attributes?

```python
import pandas as pd
import matplotlib.pyplot as plt

class DiamondAnalysis:
    def __init__(self, data_path):
        self.data = pd.read_csv(data_path)

    def task_one(self):
        # Task 1: Proportion of diamonds are between .30 and 1.08 carats.
        filtered_data = self.data[(self.data['carat'] >= 0.30) & (self.data['carat'] <= 1.08)]
        proportion = len(filtered_data) / len(self.data)
        return proportion

    def task_two(self):
        # Task 2: Diamond have equal x and y dimensions.
        equal_dimensions = self.data[self.data['x'] == self.data['y']]
        count = len(equal_dimensions)
        return count

    def task_three(self):
        # Task 3: Diamonds have a carat less than the mean carat value
        mean_carat = self.data['carat'].mean()
        less_than_mean = self.data[self.data['carat'] < mean_carat]
        count = len(less_than_mean)
        return count

    def task_four(self):
        # Task 4: Diamonds have a Premium cut or better.
        premium_or_better = self.data[self.data['cut'].isin(['Premium', 'Ideal'])]
        count = len(premium_or_better)
        return count

    def task_five(self):
        # Task 5: Diamond has the highest price per carat. And its value.
        self.data['price_per_carat'] = self.data['price'] / self.data['carat']
        max_price_per_carat = self.data['price_per_carat'].max()
        diamond = self.data[self.data['price_per_carat'] == max_price_per_carat]
        return diamond, max_price_per_carat

    def task_six(self):
        # Task 6: Bar plot of the 'cut' data.
        cut_counts = self.data['cut'].value_counts()
        cut_counts.plot(kind='bar')
        plt.xlabel('Cut')
        plt.ylabel('Frequency')
        plt.title('Frequencies of Individual Cut Values')
        plt.show()

    def task_seven(self):
        # Task 7: Boxplots of the diamond price for each cut.
        self.data.boxplot(column='price', by='cut')
        plt.title('Diamond Price for Each Cut')
        plt.suptitle('')
        plt.show()

    def task_eight(self):
        # Task 8: Histograms of 'x' and 'y' attributes.
        self.data[['x', 'y']].hist(bins=20)
        plt.suptitle('Distributions of Diamond Attributes')
        plt.show()

    def task_nine(self):
        # Task 9: Scatter plot of price vs. carat.
        plt.scatter(self.data['carat'], self.data['price'], alpha=0.5)
        plt.xlabel('Carat')
        plt.ylabel('Price')
        plt.title('Price vs. Carat')
        plt.show()

# Usage
if __name__ == "__main__":
    data_path = 'https://raw.githubusercontent.com/dakshit9228/Diamond-data/main/diamonds.csv'
    diamond_analysis = DiamondAnalysis(data_path)
    print("Task 1 Proportion:", diamond_analysis.task_one())
    print("Task 2 Count:", diamond_analysis.task_two())
    print("Task 3 Count:", diamond_analysis.task_three())
    print("Task 4 Count:", diamond_analysis.task_four())
    diamond, max_price_per_carat = diamond_analysis.task_five()
    print("Task 5 Diamond with Highest Price per Carat:", diamond)
    print("Task 5 Highest Price per Carat:", max_price_per_carat)

    diamond_analysis.task_six()
    diamond_analysis.task_seven()
    diamond_analysis.task_eight()
    diamond_analysis.task_nine()
```
